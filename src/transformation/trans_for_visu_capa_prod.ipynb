# Notebook pour la transformation des donnée de la capacité de production d'électicité pour la visualisation

```python
import pandas as pd
import numpy as np
```

```python
# import du fichier excel
data = pd.read_excel('/Users/yibtv/Espace-travail/projet_OM4A/data/raw/Data from CMP.xlsx', sheet_name=1)
# affichage
data.head(100)
```

Selection des colonnes, il s'agit des capacité disponible et installé selon les pays, technologie en fonction de la date de creation et la date de supression 

```python
# selection des colonnes necessaire
data = data[['Country', 'Technology', 'Installed Capacity', 'Availability Capacity', 'Commissioning Date', 'Decommissioing date']]
data = data.dropna()
data.head(100)
```

Conversion de MW en GW et transformation de noms de certaines colonnes

```python
# supresion des ligne sans valeur 
data.dropna()

# Conversion des valeur en type numerique
data['Installed Capacity'] = pd.to_numeric(data['Installed Capacity'], errors='coerce')
data['Availability Capacity'] = pd.to_numeric(data['Availability Capacity'], errors='coerce')

# sachant que  1GW = 1000MW alors :
data['Installed Capacity'] = data['Installed Capacity'] / 1000
data['Availability Capacity'] = data['Availability Capacity'] / 1000

# renommage de la colonne
data = data.rename(columns={
    'Installed Capacity':  'Installed Capacity (GW)',
    'Availability Capacity': 'Availability Capacity (GW)',
    'Decommissioing date': 'Decommissioning date'
    
})
# conversion de Decommisioning date en INTEGER
data['Decommissioning date'] = data['Decommissioning date'].astype(int)

data.head(100)
```

Remplacement de noms des technologies pour une conformité avec les termes de la base de donnée 

```python
# Dictionnaire de mapping pour les nouveaux noms
tech_mapping = {
    'Solar PV': 'PV Solar',
    'Nat. Gas': 'Natural Gas',
}

# Remplacement des termes dans la colonne 'Technology'
data['Technology'] = data['Technology'].replace(tech_mapping)

# Vérifier les nouvelles valeurs uniques
print(data['Technology'].unique())
```

# Chargement du resultat apres traitement dans le dossier processed

```python
# Sauvegarder dans un fichier excel
data.to_excel('/Users/yibtv/Espace-travail/projet_OM4A/data/data_visu/residual_capa_for_visu.xlsx', index=False)
print("Donnée sauvegardée ")
```
